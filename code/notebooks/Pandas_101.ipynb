## Leitura de arquivos

```python
digimons = pd.read_csv("data/digimons/digimons.csv")
print(f"(i,j): {digimons.shape}")

# save file is
digimons.to_csv("data/digimons/digimons2.csv", index=False)
digimons.head()
```

### Para ler uma coluna especifica podemos usar tres formas

```python
digimons.Digimon
digimons['Digimon']
digimons.iloc[:,1]
```

#### Para ler multiplas colunas ao mesmo tempo podemos usar o operador de python : ou arrays

```python
digimons.iloc[[0,1,2],1]
digimons.iloc[0:3,1]
```

#### Para ler linhas por nome podemos usar o loc

```python
digimons.loc[:6,'Digimon']
```

### Podemos tambem ler linhas validando uma condi√ß√£o

```python
digimons.Digimon == 'Kuramon'
```

### Para ler uma celula especifica podemos usar

```python
digimons['Digimon'][0]
```

### Para ler uma linha especifica podemos usar

```python
digimons.iloc[0]
```

#### Usando condicional para ler linhas

```python
digimons.loc[(digimons.Digimon == 'MetalGarurumon') | (digimons.Digimon == 'MetalGreymon')]
```

```python
digimons.loc[digimons.Stage.isin(['Ultimate','Mega'])]
```

```python
digimons.loc[digimons.Attribute.notnull()]
```

```python
# Gera novo DS
digimons['Lv 50 HP'].map(lambda hp: hp + digimons['Lv 50 HP'].max())
```

```python
def balance(digimon):
    digimon['Lv50 Atk'] = digimon['Lv50 Def'] + digimon['Lv50 Spd']
    return digimon

digimons.apply(balance,axis='columns')
# geram novo DS
```

```python
digimons['Digimon'] + " : " + digimons["Attribute"]
```

```python
def isBM(digimon):
    digimon['BM'] = 'BM' in digimon['Digimon']
    return digimon

digimons.apply(isBM,axis='columns')
```
