```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
digimons = pd.read_csv("data/digimons/digimons.csv") # index_col="Number"
digimons.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=digimons)
```

```python
sns.lineplot(data=digimons['Lv 50 HP'], label="HP")
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=digimons.index, y=digimons['Lv 50 HP'])
```

```python
dig = digimons.drop(['Digimon', 'Stage', 'Type', 'Attribute','Lv 50 HP','Number','Memory','Equip Slots'], axis=1)
# only the first 10 digimons
dig = dig.head(15)
plt.figure(figsize=(16,6))
sns.heatmap(data=dig, annot=True)
```

```python
sns.scatterplot(x=digimons['Lv 50 HP'], y=digimons['Lv50 Atk'], hue=digimons['Attribute'])
```

```python
sns.regplot(x=digimons['Lv 50 HP'], y=digimons['Lv50 Atk'])
```

```python
sns.lmplot(x="Lv 50 HP", y="Lv50 Atk", hue="Type", data=digimons)
```

```python
sns.swarmplot(x=digimons['Lv 50 HP'], y=digimons['Lv50 Atk'])
```

```python
sns.kdeplot(data=digimons,x='Lv50 Atk',hue="Type",fill=True)
```

```python
sns.jointplot(x=digimons['Lv 50 HP'], y=digimons['Lv50 Atk'], kind="kde")
```

```python
sns.histplot(data=digimons,x="Lv 50 HP", hue="Type")
```
